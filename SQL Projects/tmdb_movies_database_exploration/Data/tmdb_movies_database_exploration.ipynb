```python
## Importing libraries for sql analysis
import pandas as pd
import sqlite3
```

```python
cnn = sqlite3.connect('jupyter_sql_tutorial.db')

df_cred.to_sql('credits', cnn)
df_movie.to_sql('movies',cnn)
%load_ext sql

%sql sqlite:///jupyter_sql_tutorial.db
```

```python
%config SqlMagic.style ='_DEPRECATED_DEFAULT'
```

```python
## Top rated movies based on vote_count
%%sql

SELECT movies.original_title, movies.vote_average, movies.vote_count
FROM movies
ORDER BY movies.vote_average DESC,movies.vote_count DESC
LIMIT 10;
```

```python
## Average Rating by Genre
%%sql

SELECT movies.genres, AVG(movies.vote_average) AS avg_rating
FROM movies
JOIN credits ON movies.id = credits.movie_id
GROUP BY movies.genres
ORDER BY avg_rating DESC;
```

```python
## Most profitable movies
%%sql
SELECT movies.original_title,movies.profit AS profit
FROM movies
ORDER BY profit DESC
LIMIT 10;
```

```python
## Most common languages
%%sql
SELECT movies.original_language, COUNT(*) AS count_common_language
FROM movies
GROUP BY original_language
ORDER BY count_common_language DESC;
```

```python
## per year movie releases from 2000 to 2017.
%%sql
SELECT movies.year AS year, COUNT(*) AS total_movies
FROM movies
WHERE movies.year>=2000
GROUP BY year
ORDER BY year ;
```

```python
## Top directors
%%sql
SELECT credits.crew AS director, COUNT(movies.original_title) AS total_movies
FROM credits
JOIN movies ON credits.movie_id = movies.id
GROUP BY credits.crew
ORDER BY total_movies DESC
LIMIT 5;
```

we are unaware of the name of the director who made 28 films as the data is not given.

```python
##  Movies_rank by Revenue within Each Year
%%sql

SELECT
    movies.original_title,
    movies.year,
    movies.revenue,
    strftime('%Y', movies.release_date) AS release_year,
    RANK() OVER (
        PARTITION BY strftime('%Y', movies.release_date)
        ORDER BY revenue DESC
    ) AS revenue_rank
FROM movies
WHERE movies.revenue > 0;
```

```python
## Genre popularity over time
%%sql

SELECT
    m.year AS movies_year,
    m.genres,
    COUNT(*) AS movie_count
FROM movies m
JOIN credits c ON c.movie_id = m.id
WHERE m.year>=2000
GROUP BY m.year, m.genres
ORDER BY m.year, movie_count DESC;
```

```python
## Ranking Top directors by average profit
%%sql

SELECT
    dm.director,
    dm.original_title,
    dm.profit,
    AVG(dm.profit) OVER (PARTITION BY dm.director) AS avg_director_profit,
    RANK() OVER (
        PARTITION BY dm.director
        ORDER BY dm.profit DESC
    ) AS rank_within_director
FROM (
    SELECT
        c.crew AS director,
        m.original_title,
        m.revenue,
        m.budget,
        (m.revenue - m.budget) AS profit
    FROM movies m
    JOIN credits c ON c.movie_id = m.id
WHERE m.revenue > 0 AND m.budget > 0 AND (m.revenue - m.budget) > 0

) AS dm
LIMIT 5;
```

```python
##Popular movies per year
%%sql

SELECT original_title,release_year,popularity,rank
FROM (
    SELECT
        original_title,
        strftime('%Y', release_date) AS release_year,
        popularity,
        DENSE_RANK() OVER (
            PARTITION BY strftime('%Y', release_date)
            ORDER BY popularity DESC
        ) AS rank
    FROM movies
    WHERE release_date IS NOT NULL
) AS ranked
WHERE release_year >= '2000' AND rank <= 3
ORDER BY release_year, rank;
```

```python
## Average Profit per Actor and Movie Profit Rank
%%sql

SELECT
    "cast" AS actor,
    original_title,
    profit,
    ROUND(AVG(profit) OVER (PARTITION BY "cast"), 2) AS avg_actor_profit,
    RANK() OVER (PARTITION BY "cast" ORDER BY profit DESC) AS movie_profit_rank
FROM (
    SELECT
        m.original_title,
        m.revenue - m.budget AS profit,
        c."cast"
    FROM movies m
    JOIN credits c ON c.movie_id = m.id
    WHERE m.revenue > 0 AND m.budget > 0
) AS actor_profits
ORDER BY actor, movie_profit_rank
LIMIT 10;
```
